```python
# Kaggle playground competition
# Predicting defects in steel plates

import pandas as pd
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score, accuracy_score


df_train = pd.read_csv('/kaggle/input/playground-series-s4e3/train.csv')

cols = df_train.columns
target_cols = ['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains', 'Dirtiness', 'Bumps', 'Other_Faults']
feature_cols = df_train.columns.drop(target_cols + ['id'])

X_raw = df_train[feature_cols]
y_raw = df_train[target_cols]

X_train, X_test, y_train, y_test = train_test_split(X_raw, y_raw, test_size=0.2, random_state=1733, shuffle=True)

### Dirtiness predictions with decision trees ###
# Preprocessing

# Model construction
dt = DecisionTreeClassifier(max_depth=10, random_state=1733)

dt.fit(X_train, y_train['Dirtiness'])

y_dirtiness_pred = dt.predict(X_test)
acc = accuracy_score(y_test['Dirtiness'], y_dirtiness_pred)
print("Test set accuracy: {:.2f}".format(acc))

```
